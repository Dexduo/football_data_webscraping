```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
import pandas as pd
import time
```

**Iniciando o Driver**

```python
driver = webdriver.Chrome()
driver.get("https://www.transfermarkt.com.br/campeonato-brasileiro-serie-a/gesamtspielplan/wettbewerb/BRA1/saison_id/2024")
```

**Capturando dados básicos de jogos: equipes e placar**

```python
time_casa = pd.Series([])
gols_casa = pd.Series([])
gols_visitante = pd.Series([])
time_visitante = pd.Series([])
```

```python
dados_jogos = driver.find_elements(By.CLASS_NAME, "hauptlink")

indice = 0
while indice < len(driver.find_elements(By.CLASS_NAME, "hauptlink")):
    #capturando time da casa
    casa = dados_jogos[indice].find_element(By.TAG_NAME, "a")
    time_casa = pd.concat([time_casa, pd.Series([casa.accessible_name])], ignore_index=True)
    indice = indice + 1

    #capturando placar
    placar_jogo = dados_jogos[indice].find_element(By.TAG_NAME, "a")
    gols_casa = pd.concat([gols_casa, pd.Series([placar_jogo.accessible_name[0]])], ignore_index=True)
    gols_visitante = pd.concat([gols_visitante, pd.Series([placar_jogo.accessible_name[2]])], ignore_index=True)
    indice = indice + 1

    #capturando time visitante
    visitante = dados_jogos[indice].find_element(By.TAG_NAME, "a")
    time_visitante = pd.concat([time_visitante, pd.Series([visitante.accessible_name])], ignore_index=True)
    indice = indice + 1
```

```python
resultados = pd.DataFrame({
    "time_casa": time_casa,
    "gols_casa": gols_casa,
    "gols_visitante": gols_visitante,
    "time_visitante": time_visitante
})
```

```python
# pd.Series([int(resultados['gols_casa'])+int(resultados['gols_visitante'])])
resultados[-10:]
```

```python
#convertendo valor de placar para numérico e placar de jogos não realizados para NaN

resultados['gols_casa'] = pd.to_numeric(resultados['gols_casa'], errors='coerce')
resultados['gols_visitante'] = pd.to_numeric(resultados['gols_visitante'], errors='coerce')
```

```python
#criação de coluna com o total de gols da partida

resultados['gols_total'] = resultados['gols_casa'] + resultados['gols_visitante']
```

```python
#média de gols do São Paulo dentro e fora de casa

media_gols_sp_casa = resultados[resultados['time_casa'] == 'São Paulo']['gols_casa'].mean()
media_gols_sp_fora = resultados[resultados['time_visitante'] == 'São Paulo']['gols_visitante'].mean()

moda_gols_sp_casa = resultados[resultados['time_casa'] == 'São Paulo']['gols_casa'].mode()[0]
moda_gols_sp_fora = resultados[resultados['time_visitante'] == 'São Paulo']['gols_visitante'].mode()[0]

print("A MÉDIA de gols do Sâo Paulo dentro de casa é {:.2f}".format(media_gols_sp_casa))
print("A MÉDIA de gols do Sâo Paulo fora de casa é {:.2f}".format(media_gols_sp_fora))
print()
print("A MODA de gols do Sâo Paulo dentro de casa é {}".format(moda_gols_sp_casa))
print("A MODA de gols do Sâo Paulo fora de casa é {}".format(moda_gols_sp_fora))
```

```python
#analisando um duelo específico Cruzeiro x Corinthians
#veremos a media e moda dentro de casa do Cruzeiro, assim como a do Corinthians fora de casa

media_gols_cru_casa = resultados[resultados['time_casa'] == 'Bragantino']['gols_casa'].mean()
moda_gols_cru_casa = resultados[resultados['time_casa'] == 'Bragantino']['gols_casa'].mode()[0]

media_gols_cor_fora = resultados[resultados['time_visitante'] == 'Fortaleza']['gols_visitante'].mean()
moda_gols_cor_fora = resultados[resultados['time_visitante'] == 'Fortaleza']['gols_visitante'].mode()[0]

print("A MÉDIA de gols do Bragantino dentro de casa é {:.2f}".format(media_gols_cru_casa))
print("A MODA de gols do Bragantino dentro de casa é {:.2f}".format(moda_gols_cru_casa))
print()
print("A MÉDIA de gols do Fortaleza fora de casa é {:.2f}".format(media_gols_cor_fora))
print("A MODA de gols do Fortaleza fora de casa é {:.2f}".format(moda_gols_cor_fora))
```

```python
#analisando um duelo específico Cruzeiro x Corinthians
#veremos a media e moda dentro de casa do Cruzeiro, assim como a do Corinthians fora de casa

media_gols_cru_casa = resultados[resultados['time_casa'] == 'Fluminense']['gols_casa'].mean()
moda_gols_cru_casa = resultados[resultados['time_casa'] == 'Fluminense']['gols_casa'].mode()[0]

media_gols_cor_fora = resultados[resultados['time_visitante'] == 'São Paulo']['gols_visitante'].mean()
moda_gols_cor_fora = resultados[resultados['time_visitante'] == 'São Paulo']['gols_visitante'].mode()[0]

print("A MÉDIA de gols do Fluminense dentro de casa é {:.2f}".format(media_gols_cru_casa))
print("A MODA de gols do Fluminense dentro de casa é {:.2f}".format(moda_gols_cru_casa))
print()
print("A MÉDIA de gols do São Paulo fora de casa é {:.2f}".format(media_gols_cor_fora))
print("A MODA de gols do São Paulo fora de casa é {:.2f}".format(moda_gols_cor_fora))
```

Captura dos Dados dos Jogos

```python
chrome_options = Options()

prefs = {"profile.default_content_setting_values.notifications": 2}

chrome_options.add_experimental_option("prefs", prefs)
chrome_options.page_load_strategy = 'eager'
service = Service(service_args=['--read-timeout=300'])


driver = webdriver.Chrome(options=chrome_options, service=service)
pid = service.process.pid
# driver = webdriver.Chrome(options=chrome_options)
# driver = webdriver.Chrome()
driver.get("https://www.transfermarkt.com.br/campeonato-brasileiro-serie-a/gesamtspielplan/wettbewerb/BRA1/saison_id/2024")

titulo = driver.title

try:
    wait = WebDriverWait(driver, 5)
    wait.until(EC.frame_to_be_available_and_switch_to_it("sp_message_iframe_954038"))

    wait = WebDriverWait(driver, 7)
    botao = wait.until(EC.element_to_be_clickable((By.CLASS_NAME, "accept-all")))
    botao.click()

    driver.switch_to.default_content()
except:
    pass
driver.switch_to.default_content()

time.sleep(3)
```

```python
dataset_jogos = pd.DataFrame(columns=rotulos) #declaração do dataframe para acumular os dados dos jogos
```

```python
try:
    wait = WebDriverWait(driver, 7)

    botao = wait.until(EC.element_to_be_clickable((By.CLASS_NAME, "accept-all")))

    botao.click()

    driver.switch_to.default_content()
except:
    pass
```

```python
#para entrarmos na pagina do jogo
wait = WebDriverWait(driver, 7)
wait.until(EC.element_to_be_clickable((By.CLASS_NAME, "ergebnis-link")))
links_estatistica = driver.find_elements(By.CLASS_NAME, "ergebnis-link")
```

```python
tab_atual = driver.current_window_handle

jogo = 0
while jogo<len(links_estatistica):
    try:
        # time.sleep(3)
        link = links_estatistica[jogo].get_attribute('href')
        
        if(jogo%5 == 0):
            print(link, jogo)
            
        driver.switch_to.new_window('tab')
        driver.get(link)

        try:
            driver.switch_to.frame("google_ads_iframe_/58778164/transfermarkt.com.br/transfermarkt.com.br_d_interstitial_0")
            fechar_anuncio = driver.find_elements(By.CLASS_NAME, "skip")
            fechar_anuncio[0].click()
            driver.switch_to.default_content()
        except:
            pass
        driver.switch_to.default_content()

        # time.sleep(2)
        pag_est = driver.current_url.replace('index', 'statistik')

        driver.get(pag_est)

        # time.sleep(2)
        #capturando os times da partida
        times = []
        wait = WebDriverWait(driver, 5)
        wait.until(EC.element_to_be_clickable((By.CLASS_NAME, "sb-vereinslink")))
        dados = driver.find_elements(By.CLASS_NAME, "sb-vereinslink")
        for indice in range(0, len(dados)):
            times.append(dados[indice].text)


        #capturando o placar da partida
        placar = []
        wait = WebDriverWait(driver, 5)
        wait.until(EC.element_to_be_clickable((By.CLASS_NAME, "sb-endstand")))
        dados = driver.find_elements(By.CLASS_NAME, "sb-endstand")
        placar.append(dados[0].text[0])
        placar.append(dados[0].text[2])
        placar = [int(s) for s in placar]


        #catpura de dados estatísticos do jogo
        wait = WebDriverWait(driver, 5)
        wait.until(EC.element_to_be_clickable((By.CLASS_NAME, "sb-statistik-zahl")))
        dados = driver.find_elements(By.CLASS_NAME, "sb-statistik-zahl")
        lista_dados = []
        for indice in range(0, len(dados)):
            lista_dados.append(dados[indice].text)
        lista_dados = [int(s) for s in lista_dados]
        
        #concatenação de todos os dados desse jogo
        dados_jogo = times + placar + lista_dados
        
        #adição dos dados desse jogo ao dataset
        dataset_jogos.loc[len(dataset_jogos)] = pd.Series(dados_jogo).values

        
        driver.close()
        driver.switch_to.window(tab_atual)
        
    except:
        parent = psutil.Process(pid)

        for child in parent.children(recursive=True):
            try:
                child.kill()
            except psutil.NoSuchProcess:
                pass

        parent.kill()
        
        # driver.switch_to.window(tab_atual)

        chrome_options = Options()
        prefs = {"profile.default_content_setting_values.notifications": 2}
        chrome_options.add_experimental_option("prefs", prefs)
        chrome_options.page_load_strategy = 'eager'
        service = Service(service_args=['--read-timeout=300'])

        driver = webdriver.Chrome(options=chrome_options, service=service)
        pid = service.process.pid
        # driver = webdriver.Chrome(options=chrome_options)
        # driver = webdriver.Chrome()
        driver.get("https://www.transfermarkt.com.br/campeonato-brasileiro-serie-a/gesamtspielplan/wettbewerb/BRA1/saison_id/2024")

        titulo = driver.title
        tab_atual = driver.current_window_handle

        try:
            wait = WebDriverWait(driver, 5)
            wait.until(EC.frame_to_be_available_and_switch_to_it("sp_message_iframe_954038"))

            wait = WebDriverWait(driver, 7)
            botao = wait.until(EC.element_to_be_clickable((By.CLASS_NAME, "accept-all")))
            botao.click()

            driver.switch_to.default_content()
        except:
            pass
        driver.switch_to.default_content()

        time.sleep(2)

        #para entrarmos na pagina do jogo
        wait = WebDriverWait(driver, 7)
        wait.until(EC.element_to_be_clickable((By.CLASS_NAME, "ergebnis-link")))
        links_estatistica = driver.find_elements(By.CLASS_NAME, "ergebnis-link")

        jogo = jogo+1
        continue

    jogo = jogo+1
```
